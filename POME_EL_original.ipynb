```python
import random

# Initial values for AS and AD
AS = 100
AD = 100

def adjust_supply(factor):
    global AS
    AS += factor
    print(f"Aggregate Supply adjusted by {factor}. New AS = {AS}")

def adjust_demand(factor):
    global AD
    AD += factor
    print(f"Aggregate Demand adjusted by {factor}. New AD = {AD}")

def check_equilibrium():
    if AS == AD:
        print("Economy is in equilibrium!")
        return True
    elif AS > AD:
        print("Economy is in a surplus.")
    else:
        print("Economy is in a deficit.")
    return False

def play_round():
    scenarios = [
        ("Increase in consumer confidence", 10, 5),
        ("Technological advancement", 5, 10),
        ("Government spending cuts", -10, -5),
        ("Natural disaster", -5, -10),
    ]

    scenario = random.choice(scenarios)
    print(f"Scenario: {scenario[0]}")

    action = input("Choose action (1: Adjust Supply, 2: Adjust Demand): ")
    if action == '1':
        adjust_supply(scenario[1])
    elif action == '2':
        adjust_demand(scenario[2])

    return check_equilibrium()

def main():
    print("Welcome to EconBalancer!")
    rounds = 5
    for i in range(rounds):
        print(f"\n--- Round {i+1} ---")
        if play_round():
            break
    else:
        print("\nGame Over. Economy did not reach equilibrium.")

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
from tkinter import messagebox
import random

# Initial values for AS and AD
AS = 100
AD = 100

def adjust_supply(factor):
    global AS
    AS += factor
    supply_label.config(text=f"Aggregate Supply: {AS}")

def adjust_demand(factor):
    global AD
    AD += factor
    demand_label.config(text=f"Aggregate Demand: {AD}")

def check_equilibrium():
    if AS == AD:
        messagebox.showinfo("Result", "Economy is in equilibrium!")
        return True
    elif AS > AD:
        result_label.config(text="Economy is in a surplus.")
    else:
        result_label.config(text="Economy is in a deficit.")
    return False

def play_round():
    scenarios = [
        ("Increase in consumer confidence", 10, 5),
        ("Technological advancement", 5, 10),
        ("Government spending cuts", -10, -5),
        ("Natural disaster", -5, -10),
    ]
    scenario = random.choice(scenarios)
    scenario_label.config(text=f"Scenario: {scenario[0]}")
    global current_scenario
    current_scenario = scenario

def on_adjust_supply():
    adjust_supply(current_scenario[1])
    if check_equilibrium():
        disable_buttons()

def on_adjust_demand():
    adjust_demand(current_scenario[2])
    if check_equilibrium():
        disable_buttons()

def disable_buttons():
    supply_button.config(state=tk.DISABLED)
    demand_button.config(state=tk.DISABLED)

def next_round():
    supply_button.config(state=tk.NORMAL)
    demand_button.config(state=tk.NORMAL)
    play_round()

def main():
    play_round()

# GUI setup
root = tk.Tk()
root.title("EconBalancer")

scenario_label = tk.Label(root, text="Scenario: ", font=("Arial", 14))
scenario_label.pack(pady=10)

supply_label = tk.Label(root, text=f"Aggregate Supply: {AS}", font=("Arial", 12))
supply_label.pack()

demand_label = tk.Label(root, text=f"Aggregate Demand: {AD}", font=("Arial", 12))
demand_label.pack()

result_label = tk.Label(root, text="", font=("Arial", 12))
result_label.pack(pady=10)

supply_button = tk.Button(root, text="Adjust Supply", command=on_adjust_supply, font=("Arial", 12))
supply_button.pack(side=tk.LEFT, padx=20, pady=10)

demand_button = tk.Button(root, text="Adjust Demand", command=on_adjust_demand, font=("Arial", 12))
demand_button.pack(side=tk.RIGHT, padx=20, pady=10)

next_button = tk.Button(root, text="Next Round", command=next_round, font=("Arial", 12))
next_button.pack(pady=10)

main()
root.mainloop()
```
